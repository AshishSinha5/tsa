```r
#Generate Noise
noise = rnorm(10000)
noise
```

```r
ma_2 = NULL
for(i in (3:10000)){
    ma_2[i] = noise[i] + 0.5*noise[i-1] + 0.5*noise[i-2]
}
ma_process = ma_2[3:10000]
ma_process = ts(ma_process)
par(mfrow = c(2,1))
plot(ma_process, main='A moving average process of order 2', ylab=' ', col='blue')
(acf(ma_process, main='Correlogram of a moving average process of order 2'))
```
