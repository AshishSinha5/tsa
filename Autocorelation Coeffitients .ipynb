### Generating a random walk

```r
x = NULL
x[1] = 0
for(i in 2:10000){
    x[i] = x[i-1] + rnorm(1)
}
```

```r
random_walk = ts(x)
```

```r
plot(random_walk,main = "Random Walk", xlab = "Days", col = "blue",ylab ="",lw = 1)
```

```r
(acf(random_walk))
```

We can see that there is high correlation even in we increase the lag, indicating that there is no stationarity in the time series data

### Removing The trend

```r
plot((diff(random_walk)))
```

Looks like white noise

```r
(acf(diff(random_walk)))
```
